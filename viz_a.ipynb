Learning more about people through data visualization of profile information on the online dating website, OK Cupid.

```python
# Poetry dependencies
# pandas matplotlib jupyterlab seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Import the Data

```python
data = pd.read_csv('./okcupid_profiles.csv')
# data.head(5)
```

```python
# ignore "essay" text columns
# profiles = okcupid_profiles
okcupid_profiles = data.iloc[:,0:21]
```

```python
# breakdown by age
plt.figure(figsize=(10, 5))
sns.catplot(y='age', data = okcupid_profiles, kind='count', palette='pastel', hue='sex', height=10, aspect=1) # make plot 10 units high
plt.yticks(rotation = 25)
plt.title('Age Distribution')
sns.set()
# sns.set(style='whitegrid')
sns.set_style('darkgrid')
sns.set(font='serif')
plt.show() 
```

A surprising broad range of ages are engaging in online dating.
- For males we assess that males in their mid-20s have the higest membership.
- Females have a similar massing in the mid-20s. 
- From age 17 - 52 males outnumber females. Females appear to close or overcome the gap between the ages of 56 - 69 years old.

```python
#  breakdown by ethnicity
plt.figure(figsize=(10, 5))
sns.countplot(y='ethnicity', data=data, hue="sex",
palette='pastel',
order = data['ethnicity'].value_counts().iloc[:10].index)
plt.show()
```

Reviewing the ethnic breakdown of membership, we can observe the following:
- Whites are the vast majority.
- Males outnumber females regardless of ethnic group.
- Seeing the small number of Indians, Asians, and Pacific Islanders could indicate a different way that they may view dating. Perhaps sticking to more traditional relationships.
- Other is suspect. Expect that people use other as a cover.

```python
# location data
plt.figure(figsize=(10, 5))
sns.countplot(y='location', data=data, hue="sex",
palette='Paired', 
order = data['location'].value_counts().iloc[:10].index)
plt.show() 
```

People in California are lonely.
For some reason San Francisco, California has the most members int he USA.

```python
# must love dogs
plt.figure(figsize=(10, 5))
sns.countplot(y = 'pets', data=okcupid_profiles, hue = 'sex', palette = 'PuRd', order = okcupid_profiles['pets'].value_counts().iloc[:].index)
sns.set()
# sns.set(style='whitegrid')
sns.set_style('darkgrid')
sns.set(font='serif')
plt.show() 
```

People appear to desire people how like pets.
- People that dislike dogs are bad people.
- More people like and have dogs than cats. 

```python
# breakdown by age
plt.figure(figsize=(10, 5))
sns.catplot(x='job', data = okcupid_profiles, kind='count', palette='pastel', hue='sex', height=8, aspect=3) # make plot 10 units high
plt.xticks(rotation = 25)
plt.title('Career Field')
sns.set()
```

There is a broad spread of career fields.
- The highest participation for was among artists, sales and marketing folks, and healthcare workers.
- Women continue to outnumber males in many of the traditionally female career fields.

```python
# show me the money
plt.figure(figsize=(10, 5))
okcupid_profiles_income = okcupid_profiles[okcupid_profiles.income != -1]
sns.countplot(y = 'income', data=okcupid_profiles_income, hue='sex', palette = 'crest')
sns.despine() # removes spines from right and upper portion of plot
plt.show()
```

This income scale indicates:
- The number of males at each income level nearly doubles the females.
- The majority of members earn somewhere near $20K a year.
- It is likely that men lie about how much they make and women understate their income. 

Most people that participate in online dating on OK Cupid are in the mid-20s.
This community is predominately male and white.
Most members like pets with dogs being the pet of choice.
OK Cupid appears to be most active in California among artists and sales people.
