```python
data = pd.read_csv('./okcupid_profiles.csv')
data.head(5)
```

```python
# ignore "essay" text columns
# profiles = okcupid_profiles
okcupid_profiles = data.iloc[:,0:21]
```

```python
# breakdown by age
plt.figure(figsize=(10, 5))
sns.catplot(y='age', data = okcupid_profiles, kind='count', palette='pastel', hue='sex', height=10, aspect=1) # make plot 10 units high
plt.yticks(rotation = 25)
plt.title('Age Distribution')
sns.set()
# sns.set(style='whitegrid')
sns.set_style('darkgrid')
sns.set(font='serif')
plt.show() 
```

```python
df=(data.groupby('ethnicity') [['sex']].count().sort_values(by='sex', ascending=False))
df['% of subjects']=(df['sex']/df['sex'].sum())*100
df=df.rename(columns= {'sex':'numer of subjects'})
```

```python
#  breakdown by ethnicity
plt.figure(figsize=(10, 5))
sns.countplot(y='ethnicity', data=data, hue="sex",
palette='pastel',
order = data['ethnicity'].value_counts().iloc[:10].index)
plt.show()
```

```python
#  what's your sign
plt.figure(figsize=(10, 5))
sns.countplot(y='sign', data=data, hue="sex",
palette='tab10',
# palette='rocket_r',
order = data['sign'].value_counts().iloc[:10].index)
# sns.color_palette("rocket_r", as_cmap=True)
plt.show()
```

```python
# location data
plt.figure(figsize=(10, 5))
sns.countplot(y='location', data=data, hue="sex",
palette='Paired', 
order = data['location'].value_counts().iloc[:10].index)
plt.show() 
```

```python
# must love dogs
plt.figure(figsize=(10, 5))
sns.countplot(y = 'pets', data=okcupid_profiles, hue = 'sex', palette = 'PuRd', order = okcupid_profiles['pets'].value_counts().iloc[:].index)
sns.set()
# sns.set(style='whitegrid')
sns.set_style('darkgrid')
sns.set(font='serif')
plt.show() 
```

```python
# breakdown by age
plt.figure(figsize=(10, 5))
sns.catplot(x='job', data = okcupid_profiles, kind='count', palette='pastel', hue='sex', height=8, aspect=3) # make plot 10 units high
plt.xticks(rotation = 25)
plt.title('Career Field')
sns.set()
```

```python
# show me the money
plt.figure(figsize=(10, 5))
okcupid_profiles_income = okcupid_profiles[okcupid_profiles.income != -1]
sns.countplot(y = 'income', data=okcupid_profiles_income, hue='sex', palette = 'crest')
sns.despine() # removes spines from right and upper portion of plot
plt.show()
```
